## Import dependencies

```python
from simple_dqn import Agent
import numpy as np
from racing_car_env import RacingCarEnv
```

```python
env = RacingCarEnv()
states = env.observation_space.shape[0]
actions_n = env.action_space.n
```

```python
agent = Agent(alpha=0.0005, gamma=0.99, epsilon=1.0, input_dims=states, n_actions=actions_n, batch_size=64, epsiolon_end=0.01)
```

```python
for i in range(100):
    done = False
    score = 0
    observation = env.reset()
    while not done:
        action = agent.choosen_action(observation)
        new_observation, reward, done, _ = env.step(action)
        score += reward
        agent.rememeber(observation, action, reward, new_observation, done)
        observation = new_observation
        agent.learn()

    print(f"Epoch {i}, Score {score}")

agent.save_model()
```
